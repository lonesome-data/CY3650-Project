## Objective

Scrape novels from [Project Gutenberg](https://www.gutenberg.org/), extract text from HTML, tokenize, remove stopwords, and visualize word frequency distributions. Demonstrates the core NLP pipeline applicable to any unstructured text corpus.


## Setup


```python
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import RegexpTokenizer
import nltk
import matplotlib.pyplot as plt
import seaborn as sns

nltk.download("stopwords", quiet=True)
sns.set()
```

## Pipeline

The full NLP pipeline in four steps:
1. **Collect** — HTTP GET from Project Gutenberg
2. **Extract** — BeautifulSoup parses HTML to raw text
3. **Tokenize** — regex tokenizer splits text into lowercase words
4. **Filter** — remove English stopwords, plot frequency distribution


## 1. Collect — Fetch the HTML


```python
url = "https://www.gutenberg.org/files/8649/8649-h/8649-h.htm"
r = requests.get(url)
print(f"Status: {r.status_code}  |  Content length: {len(r.text):,} chars")
```

## 2. Extract — HTML to Text


```python
soup = BeautifulSoup(r.text, "html.parser")
print(f"Title: {soup.title.string}")

text = soup.get_text()
print(f"Extracted text: {len(text):,} characters")
```

## 3. Tokenize

A regular expression `\w+` matches sequences of alphanumeric characters — effectively splitting text into words while discarding punctuation.


```python
tokenizer = RegexpTokenizer(r"\w+")
tokens = tokenizer.tokenize(text)
words = [w.lower() for w in tokens]

print(f"Total tokens: {len(words):,}")
print(f"First 15: {words[:15]}")
```

## 4. Filter Stopwords and Plot


```python
sw = set(nltk.corpus.stopwords.words("english"))
words_filtered = [w for w in words if w not in sw]

print(f"After stopword removal: {len(words_filtered):,} words ({len(words) - len(words_filtered):,} removed)")

freqdist = nltk.FreqDist(words_filtered)
freqdist.plot(25, title="Indian Tales — Top 25 Words");
```

## Generalized Pipeline

Wrap the pipeline into a reusable function and apply to other Gutenberg texts.


```python
def plot_word_freq(url, n=25):
    """Fetch a Project Gutenberg URL and plot the top-n word frequencies."""
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    text = soup.get_text()

    tokenizer = RegexpTokenizer(r"\w+")
    words = [w.lower() for w in tokenizer.tokenize(text)]

    sw = set(nltk.corpus.stopwords.words("english"))
    words_filtered = [w for w in words if w not in sw]

    freqdist = nltk.FreqDist(words_filtered)
    freqdist.plot(n)
```

### War and Peace — Tolstoy


```python
plot_word_freq("https://www.gutenberg.org/files/2600/2600-h/2600-h.htm")
```

### Ulysses — Joyce


```python
plot_word_freq("https://www.gutenberg.org/files/4300/4300-h/4300-h.htm")
```

### King James Bible


```python
plot_word_freq("https://www.gutenberg.org/files/10/10-h/10-h.htm")
```

## Takeaways

- The NLP pipeline (collect → extract → tokenize → filter → analyze) applies to any text source, not just novels
- Stopword removal is critical — without it, function words (`the`, `and`, `of`) dominate and obscure content words
- Word frequency distributions follow Zipf's law: a small number of words account for a disproportionate share of all occurrences
- The `plot_word_freq()` function generalizes to any Project Gutenberg URL, enabling rapid comparative analysis across texts
